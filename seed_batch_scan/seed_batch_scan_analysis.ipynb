# Seed Batch Scan Analysis

Грузит батч CSV, даёт фильтры по `ask`, market/expected overpay и разнице между ними, потом печатает результат.

```python
from pathlib import Path

import pandas as pd
```

```python
RUN_DIR = Path.cwd()
CSV_PATH = RUN_DIR / "data" / "seed_batch_scan_matches.csv"
CSV_PATH
```

```python
df = pd.read_csv(CSV_PATH)
df.shape
```

```python
MAX_ASK = 100
MAX_MARKET_OVERPAY_MULT = None
MIN_EXPECTED_OVERPAY_MULT = None
MIN_OVERPAY_GAP_MULT = None
MIN_EDGE_RATIO = 0.3
MIN_CONFIDENCE = None
QUERY_SUBSTRING = ""
ITEM_SUBSTRING = ""
```

```python
filtered = df.copy()

if MAX_ASK is not None:
    filtered = filtered[filtered["ask"] <= MAX_ASK]
if MAX_MARKET_OVERPAY_MULT is not None:
    filtered = filtered[filtered["market_overpay_mult"] <= MAX_MARKET_OVERPAY_MULT]
if MIN_EXPECTED_OVERPAY_MULT is not None:
    filtered = filtered[filtered["expected_overpay_mult"] >= MIN_EXPECTED_OVERPAY_MULT]
if MIN_OVERPAY_GAP_MULT is not None:
    filtered = filtered[filtered["overpay_gap_mult"] >= MIN_OVERPAY_GAP_MULT]
if MIN_EDGE_RATIO is not None:
    filtered = filtered[filtered["edge_ratio"] >= MIN_EDGE_RATIO]
if MIN_CONFIDENCE is not None:
    filtered = filtered[filtered["overall_confidence"] >= MIN_CONFIDENCE]
if QUERY_SUBSTRING:
    filtered = filtered[filtered["query_name"].str.contains(QUERY_SUBSTRING, case=False, na=False)]
if ITEM_SUBSTRING:
    filtered = filtered[filtered["market_hash_name"].str.contains(ITEM_SUBSTRING, case=False, na=False)]

filtered = filtered.sort_values(["overpay_gap_mult", "edge_abs", "ask"], ascending=[False, False, True], na_position="last").reset_index(drop=True)
filtered.shape
```

```python
show_cols = [
    col for col in [
        "query_name",
        "market_hash_name",
        "paint_seed",
        "float_value",
        "ask",
        "base_ask_mean_10",
        "market_overpay_mult",
        "expected_overpay_mult",
        "overpay_gap_mult",
        "edge_abs",
        "edge_ratio",
        "overall_confidence",
        "seed_confidence",
        "listing_id",
    ] if col in filtered.columns
]
filtered[show_cols].sort_values("edge_ratio", ascending = False).head(100)
```
